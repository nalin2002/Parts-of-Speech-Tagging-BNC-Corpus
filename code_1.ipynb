```python
import os
import sys
from xml.etree import ElementTree as ET

#---------------------------------------------------------------------------------------------

word_frequency = dict()
tag_frequency = dict()
total_frequency = dict()
counter = 0

#---------------------------------------------------------------------------------------------

def add_c5(x):
   if x in tag_frequency:
       tag_frequency[x] += 1
   else:
       tag_frequency[x] = 1
        
def add_word(x):
   if x in word_frequency:
       word_frequency[x] += 1
   else:
       word_frequency[x] = 1

def add_total(x):
   if x.attrib['hw']+"_"+x.attrib['c5'] in total_frequency:
       total_frequency[x.attrib['hw']+"_"+x.attrib['c5']] += 1
   else:
       total_frequency[x.attrib['hw']+"_"+x.attrib['c5']] = 1

#---------------------------------------------------------------------------------------------

directory = '/home/daddy_yankee/Desktop/csn371/csn371_assignment/Train-corpus'

for filename in os.listdir(directory):
   complete_file_path = os.path.join(directory, filename)
   for file in os.listdir(complete_file_path):
       full_file = os.path.abspath(os.path.join(complete_file_path, file))
       mytree = ET.parse(full_file)
       myroot = mytree.getroot()
       original_stdout = sys.stdout
       for x in myroot.findall('./w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/*/*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/*/*/*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/*/*/*/*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./*/*/*/*/*/*/*/*/*/*/w'):
           add_c5(x.attrib['c5'])
           add_word(x.attrib['hw'])
           add_total(x)
           counter += 1
       for x in myroot.findall('./c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/*/*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/*/*/*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/*/*/*/*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/*/*/*/*/*/*/c'):
           add_c5(x.attrib['c5'])
       for x in myroot.findall('./*/*/*/*/*/*/*/*/*/*/c'):
           add_c5(x.attrib['c5'])
            
#--------------------------------------------------------------------------------------------

original_stdout = sys.stdout
    
sorted_word_frequency = sorted(word_frequency.items(), key=lambda kv: kv[1], reverse=True)
with open('word.txt', "w", encoding="utf-8") as f:
   sys.stdout = f
   for i in range(len(sorted_word_frequency)):
       print(sorted_word_frequency[i])
    
sorted_tag_frequency = sorted(tag_frequency.items(), key=lambda kv: kv[1], reverse=True)
with open('tag.txt', "w", encoding="utf-8") as f:
   sys.stdout = f
   for i in range(len(sorted_tag_frequency)):
       print(sorted_tag_frequency[i])
    
sorted_total_frequency = sorted(total_frequency.items(), key=lambda kv: kv[1], reverse=True)
with open('word_and_tag.txt', "w", encoding="utf-8") as f:
   sys.stdout = f
   for i in range(len(sorted_total_frequency)):
       print(sorted_total_frequency[i])
    
sys.stdout = original_stdout

#---------------------------------------------------------------------------------------------

print("The top 10 frequently used words are listed as follows :\n")
for i in range(10):
   print(sorted_word_frequency[i])
print("\n")

print("The top 10 frequently used tags are listed as follows :\n")
for i in range(10):
   print(sorted_tag_frequency[i])
print("\n")

print("Total number of words :", counter)
print("\n")

#---------------------------------------------------------------------------------------------
```
